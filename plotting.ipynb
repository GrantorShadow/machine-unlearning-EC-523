```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
s = "general-report-"
df_5s_15r_gm = pd.read_csv(s+"5s-15r-5sl-gm.csv")
df_5s_15r_pareto = pd.read_csv(s+"5s-15r-5sl-pareto.csv")
df_5s_20r_gm = pd.read_csv(s+"5s-20r-5sl-gm.csv")
df_5s_20r_pareto = pd.read_csv(s+"5s-20r-5sl-pareto.csv")
df_5s_25r_gm = pd.read_csv(s+"5s-25r-5sl-gm.csv")
df_5s_25r_pareto = pd.read_csv(s+"5s-25r-5sl-pareto.csv")

df_10s_15r_gm = pd.read_csv(s+"10s-15r-5sl-gm.csv")
df_10s_15r_pareto = pd.read_csv(s+"10s-15r-5sl-pareto.csv")
df_10s_20r_gm = pd.read_csv(s+"10s-20r-5sl-gm.csv")
df_10s_20r_pareto = pd.read_csv(s+"10s-20r-5sl-pareto.csv")
df_10s_25r_gm = pd.read_csv(s+"10s-25r-5sl-gm.csv")
df_10s_25r_pareto = pd.read_csv(s+"10s-25r-5sl-pareto.csv")

df_20s_15r_gm = pd.read_csv(s+"20s-15r-5sl-gm.csv")
df_20s_15r_pareto = pd.read_csv(s+"20s-15r-5sl-pareto.csv")
df_20s_20r_gm = pd.read_csv(s+"20s-20r-5sl-gm.csv")
df_20s_20r_pareto = pd.read_csv(s+"20s-20r-5sl-pareto.csv")
df_20s_25r_gm = pd.read_csv(s+"20s-25r-5sl-gm.csv")
df_20s_25r_pareto = pd.read_csv(s+"20s-25r-5sl-pareto.csv")
```

```python
time_df_5s_15r_gm = sum(df_5s_15r_gm['retraining_time'])
time_df_5s_15r_pareto = sum(df_5s_15r_pareto['retraining_time'])
time_df_5s_20r_gm = sum(df_5s_20r_gm['retraining_time'])
time_df_5s_20r_pareto = sum(df_5s_20r_pareto['retraining_time'])
time_df_5s_25r_gm = sum(df_5s_25r_gm['retraining_time'])
time_df_5s_25r_pareto = sum(df_5s_25r_pareto['retraining_time'])

time_df_10s_15r_gm = sum(df_10s_15r_gm['retraining_time'])
time_df_10s_15r_pareto = sum(df_10s_15r_pareto['retraining_time'])
time_df_10s_20r_gm = sum(df_10s_20r_gm['retraining_time'])
time_df_10s_20r_pareto = sum(df_10s_20r_pareto['retraining_time'])
time_df_10s_25r_gm = sum(df_10s_25r_gm['retraining_time'])
time_df_10s_25r_pareto = sum(df_10s_25r_pareto['retraining_time'])

time_df_20s_15r_gm = sum(df_20s_15r_gm['retraining_time'])
time_df_20s_15r_pareto = sum(df_20s_15r_pareto['retraining_time'])
time_df_20s_20r_gm = sum(df_20s_20r_gm['retraining_time'])
time_df_20s_20r_pareto = sum(df_20s_20r_pareto['retraining_time'])
time_df_20s_25r_gm = sum(df_20s_25r_gm['retraining_time'])
time_df_20s_25r_pareto = sum(df_20s_25r_pareto['retraining_time'])
```

```python
gm_line_5s = [time_df_5s_15r_gm,
              time_df_5s_20r_gm,
              time_df_5s_25r_gm]

gm_line_10s = [time_df_10s_15r_gm,
              time_df_10s_20r_gm,
              time_df_10s_25r_gm]

gm_line_20s = [time_df_20s_15r_gm,
              time_df_20s_20r_gm,
              time_df_20s_25r_gm]

pareto_line_5s = [time_df_5s_15r_pareto,
              time_df_5s_20r_pareto,
              time_df_5s_25r_pareto]

pareto_line_10s = [time_df_10s_15r_pareto,
              time_df_10s_20r_pareto,
              time_df_10s_25r_pareto]

pareto_line_20s = [time_df_20s_15r_pareto,
                   time_df_20s_20r_pareto,
                   time_df_20s_25r_pareto]

x_values = [15, 20, 25]
plt.figure(figsize=(8, 6))
plt.plot(x_values, gm_line_5s, label='greedy (proposed), 5 shards', linestyle='dashed', marker='o')
plt.plot(x_values, gm_line_10s, label='greedy (proposed), 10 shards', linestyle='dashed', marker='o')
plt.plot(x_values, gm_line_20s, label='greedy (proposed), 20 shards', linestyle='dashed', marker='o')

plt.plot(x_values, pareto_line_5s, label='pareto (SISA), 5 shards', linestyle='dashdot', marker='o')
plt.plot(x_values, pareto_line_10s, label='pareto (SISA), 10 shards', linestyle='dashdot', marker='o')
plt.plot([15, 20, 25], pareto_line_20s, label='pareto (SISA), 20 shards', linestyle='dashdot', marker='o')

plt.xlabel('Number of Requests')
plt.ylabel('Sum of retraining time (s)') 
plt.title("Distribution-aware sharding - greedy (proposed) vs. SISA")
plt.legend()
plt.grid(x_values)
plt.xticks(x_values)
plt.show()
```
